```python
# Load the CSVs into pandas DataFrames
train = pd.read_csv("/content/drive/MyDrive/samsum-dataset-text-summarization/samsum-train.csv")
test = pd.read_csv("/content/drive/MyDrive/samsum-dataset-text-summarization/samsum-test.csv")
val = pd.read_csv("/content/drive/MyDrive/samsum-dataset-text-summarization/samsum-validation.csv")
```

```python
# Convert pandas DataFrames to Hugging Face Datasets
train_dataset = Dataset.from_pandas(train)
test_dataset = Dataset.from_pandas(test)
val_dataset = Dataset.from_pandas(val)
```

```python
# Convert to Hugging Face Dataset
# Combine into a DatasetDict

from datasets import Dataset, DatasetDict

dataset_samsum = DatasetDict({
    'train': train_dataset,
    'test': test_dataset,
    'validation': val_dataset
})
```

```python
def convert_examples_to_features(example_batch):
    # Ensure all dialogues and summaries are strings and not None
    dialogues = [str(d) if d is not None else "" for d in example_batch["dialogue"]]
    summaries = [str(s) if s is not None else "" for s in example_batch["summary"]]

    # Tokenize inputs (dialogues)
    input_encodings = tokenizer(
        dialogues,
        max_length=1024,
        truncation=True,
        padding="max_length"
    )

    # Tokenize targets (summaries)
    target_encodings = tokenizer(
        summaries,
        max_length=128,
        truncation=True,
        padding="max_length"
    )

    return {
        "input_ids": input_encodings["input_ids"],
        "attention_mask": input_encodings["attention_mask"],
        "labels": target_encodings["input_ids"]
    }
```

```python
# Apply to dataset
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched=True)
dataset_samsum_pt["train"]
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    output_dir='pegasus-samsum',
    num_train_epochs=1,
    warmup_steps=500,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    weight_decay=0.01,
    logging_steps=10,
    eval_strategy='steps',
    eval_steps=500,
    save_steps=1e6,
    gradient_accumulation_steps=16
)


    # output_dir="./results",
    # run_name="pegasus_samsum_run1",
```

```python
trainer = Trainer(model=model_pegasus, args=trainer_args,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt["train"],
                  eval_dataset=dataset_samsum_pt["validation"])
```

```python
trainer.train()
```

```python
# Utility to generate batches
def generate_batch_sized_chunks(elements, batch_size):
    """Yield successive batch-sized chunks from a list."""
    for i in range(0, len(elements), batch_size):
        yield elements[i: i + batch_size]

# Compute ROUGE on the test set
def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                                batch_size=16, device="cuda",
                                column_text="dialogue",
                                column_summary="summary"):
    text_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    summary_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for text_batch, ref_batch in tqdm(zip(text_batches, summary_batches), total=len(text_batches), desc="Evaluating"):
        inputs = tokenizer(
            text_batch,
            max_length=1024,
            truncation=True,
            padding="max_length",
            return_tensors="pt"
        )
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = model.generate(
                input_ids=inputs["input_ids"],
                attention_mask=inputs["attention_mask"],
                length_penalty=0.8,
                num_beams=8,
                max_length=128
            )

        decoded_preds = tokenizer.batch_decode(
            outputs,
            skip_special_tokens=True,
            clean_up_tokenization_spaces=True
        )

        # Wrap references in list-of-lists as expected by ROUGE
        metric.add_batch(predictions=decoded_preds, references=[[ref] for ref in ref_batch])

    return metric.compute()
```

```python
# Load ROUGE metric
rouge_metric = load_metric('rouge')
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
```

```python
score = calculate_metric_on_test_ds(
    dataset=dataset_samsum['test'],
    metric=rouge_metric,
    model=trainer.model,
    tokenizer=tokenizer,
    batch_size=2,
    column_text='dialogue',
    column_summary='summary'
)
```

```python
# print ROUGE scores
rouge_dict = {rn: score[rn].mid.fmeasure for rn in rouge_names}
print("\n ROUGE Scores:")
print(pd.DataFrame(rouge_dict, index=['pegasus']))
```
